```python
import pandas as pd
import numpy as np


```

# load CSV

```python
r_cols = ["user_id", "movie_id", "rating"]   
ratings = pd.read_csv("movies_data.txt", sep= '\t', names=r_cols, usecols=range(3))
ratings.head(10)
```

```python
movieProperties = ratings.groupby('movie_id').agg({'rating': [np.size, np.mean]})
movieProperties.head(10)

```

```python
movieNumRatings = pd.DataFrame(movieProperties['rating']['size'])
movieNumRatings.head(10)
```

```python
movieNormalizedRatings = movieNumRatings.apply(lambda x : (x - np.min(x)) / (np.max(x) - np.min(x)))
print(len(movieNormalizedRatings))
movieNormalizedRatings.head()
```

```python
moviesDict = []

with open (r"./movies_item.txt") as f:
    for line in f:
        fields = line.rstrip("\n").split('|')
        movieId = int(fields[0])
        movieName = fields[1]
        genres = fields[5:25]
        genres = map(int, genres)
        moviesDict[movieId] = (movieName, genres, movieNormalizedRatings.loc[movieId].get('size'), movieProperties.loc[movieId].rating.get('mean'))    
        # print(movieNormalizedRatings.loc[moviesId].get('size'))
        
```
